### Exercise 1 

Two unbiased dice are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7

We could use a mathematical equation to solve this or we can run the experiment many times, and then count the number of times where the score is even or greater than 7

-Run the experiment 1000 times (roll 2 dice 1000 times, and sum the result)

-Keep track of the number of times that the sum was either greater than 7 or an even number

-Divide the number from step 2 by the number of iterations (1000)


```python
#import libraries
import pandas as pd
import numpy as np
import random

#Realizamos una funcion

def roll_the_dice (n_simulations = 1000):
    count = 0
    #Vamos a usar for para iterar

    for i in range (n_simulations):

        die1 = random.randint(1,6) # The randint() method returns an integer number selected element from the specified range.
        die2 = random.randint(1,6) # The randint() method returns an integer number selected element from the specified range.

        # suma
        score = die1 + die2

        if score %2 == 0 or score >7:
           count += 1

    return count/n_simulations

string= 'La probabilidad que sea un par o mayor que 7 es'
print (string,np.round(roll_the_dice()*100,2),'%' ) #numpy.roundÂ¶ ... Round an array to the given number of decimals



```

```python
cantExp=1000 #count experiment
exito=0

for i in range(1,cantExp):
    sum= random.randint(1,6) + random.randint(1,6)
    if sum %2 == 0 or sum >7:
           exito += 1

prob= exito/cantExp

print('probabilidad' +'     '+ str(prob*100))
```

### Exercise 2

A box contains 10 white balls, 20 reds and 30 greens. Draw 5 balls with replacement. What is the probability that:

a. 3 white or 2 red

b. All 5 are the same color


We will use the same system here. We will pick our balls during each round and count the number of times that:

a. 3 white or 2 red 

b. All 5 are the same color.

```python
#import libraries
import pandas as pd
import numpy as np
import random

d = ['white']*10 +['red']*20 + ['green']*30

#inicializamos la funcion
n_simulations = 1000
countA = 0 
countB = 0

for i in range(n_simulations):
    
    choosen = np.random.choice (d,5)
    
    #find the number of each that we picked
    white =np.sum([choosen == 'white'])
    red= np.sum([choosen =='red'])
    green= np.sum([choosen =='green'])
    
    #Keep track if the combination met the criteria
    if white == 3 and red == 2:
       countA += 1
    
    if red == 5 or white == 5 or green ==5:
       countB +=1


print('The probability of 3 white and 2 red is:', round(countA/n_simulations*100,2),'%')
print('The probability of all the same color is:', round(countB/n_simulations*100,2),'%')
```
